```python
def create_hist(table, column):
  fig, ax = plt.subplots()
  x = table[column]
  ax.hist(x)
  ax.set_title(f'Histogram of {column}')

  plt.show()
```

```python
counts = iris_df['class'].value_counts()

plt.barh(counts.index, counts.values)
plt.title('Observation by Class')
```

```python
iris_virg = iris_df[iris_df['class'].str.lower() == "iris-virginica"]
iris_vers = iris_df[iris_df['class'].str.lower() == "iris-versicolor"]
iris_seto = iris_df[iris_df['class'].str.lower() == "iris-setosa"]
```

```python
# get summary for all variable per class
def create_histograms(dataframe):

  fig, ((ax0,ax1), (ax2, ax3)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

  # Histogram for Sepal Length
  ax0.hist(dataframe['sepal_length'])
  ax0.set_title('Sepal Length')
  ax0.set_ylabel('Count')

  # Histogram for Sepal Width
  ax1.hist(dataframe['sepal_width'])
  ax1.set_title('Sepal Width')
  ax1.set_ylabel('Count')


  # Histogram for Petal Length
  ax2.hist(dataframe['petal_length'])
  ax2.set_title('Petal Length')
  ax2.set_ylabel('Count')

  # Histogram for Petal Width
  ax3.hist(dataframe['petal_width'])
  ax3.set_title('Petal Width')
  ax3.set_ylabel('Count')

  print('Info')
  dataframe.info()
  print('\n')

  print('Histograms')
  fig.tight_layout()
  plt.show()
```

```python
create_histograms(iris_virg)
```

```python
create_histograms(iris_vers)
```

```python
create_histograms(iris_seto)
```

```python
def compare_variables(dataframe1, dataframe2, dataframe1_title, dataframe2_title):

  fig, ((ax0,ax1), (ax2, ax3)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

  # Histogram for Sepal Length
  ax0.hist(dataframe1['sepal_length'],label = 'dataframe1_title', alpha=0.7)
  ax0.hist(dataframe2['sepal_length'], label = 'dataframe2_title', alpha=0.7)
  ax0.set_title(f'Sepal Lengths of {dataframe1_title} and {dataframe2_title}')
  ax0.set_ylabel('Count')


  # Histogram for Sepal Length
  ax1.hist(dataframe1['sepal_width'], label = 'dataframe1_title', alpha=0.7)
  ax1.hist(dataframe2['sepal_width'], label = 'dataframe2_title', alpha=0.7)
  ax1.set_title(f'Sepal Width of {dataframe1_title} and {dataframe2_title}')
  ax1.set_ylabel('Count')

  # Histogram for Sepal Length
  ax2.hist(dataframe1['petal_length'], label = 'dataframe1_title', alpha=0.7)
  ax2.hist(dataframe2['petal_length'], label = 'dataframe2_title', alpha=0.7)
  ax2.set_title(f'Sepal Width of {dataframe1_title} and {dataframe2_title}')
  ax2.set_ylabel('Count')

  # Histogram for Sepal Length
  ax3.hist(dataframe1['petal_width'], label = 'dataframe1_title', alpha=0.7)
  ax3.hist(dataframe2['petal_width'], label = 'dataframe2_title', alpha=0.7)
  ax3.set_title(f'Sepal Width of {dataframe1_title} and {dataframe2_title}')
  ax3.set_ylabel('Count')



  fig.tight_layout()
  plt.show()
```

```python
compare_variables(iris_virg, iris_vers, 'Iris-virginica', 'Iris-versicolor')
```

```python
compare_variables(iris_vers, iris_seto, 'Iris-Versicolor', 'Iris-Setosa')
```

```python
compare_variables(iris_virg, iris_seto, 'Iris-Virginica', 'Iris-Setosa')
```
