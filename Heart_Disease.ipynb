```python
# pip install ucimlrepo
from ucimlrepo import fetch_ucirepo
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# fetch dataset
heart_disease = fetch_ucirepo(id=45)
data = heart_disease.data.original
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
fig, ax = plt.subplots()

x = df['age']

ax.hist(x)
ax.set_title('Histogram of age')

plt.show()
```

```python
sex_count = df['sex'].value_counts()

fig, ax = plt.subplots()
barchart = ax.barh(sex_count.index, sex_count.values)
ax.set_title('Number of Observation by sex')
ax.spines[:].set_visible(False)

for bar in barchart:
  plt.text(bar.get_width()  + 10, bar.get_y() + bar.get_height()/2,
           f'{bar.get_width():.0f}',ha='center', va='bottom', fontsize=12)

plt.xticks([])
plt.grid(visible = False)
plt.yticks([0,1])
plt.show()
```

```python
sex_0 = df[df['sex'] == 1]
sex_1 = df[df['sex'] == 0]


fig, ax = plt.subplots()

ax.hist(sex_0['age'], label='Sex 0', alpha = 0.7)
ax.hist(sex_1['age'], label='Sex 1', alpha = 0.7)
ax.set_title('Age comparison between Sex 0 and Sex 1')

plt.legend()
plt.show()
```
